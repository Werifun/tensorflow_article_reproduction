```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 使用numpy生成200个随机点
x_data = np.linspace(-0.5,0.5,200)[:, np.newaxis]
noise = np.random_normal(0, 0.02, x_data.shape)
print('x_data shape', x_data.shape)
y_data = np.square(x_data) + noise

```

```python
# 定义两个placeholder
x = tf.placeholder(tf.float32, [None, 1])
y = tf.placeholder(tf.float32, [None, 1])

# 定义神经网络中间层
Weights_L1 = tf.Variable(tf.random_normal([1, 10]))  # 因为输入成是1个神经元
biases_L1 = tf.Variable(tf.zeros([1, 10]))
Wx_plus_b_L1 = tf.matmul(x, Weights_L1) + biases_L1  #因为是矩阵乘法，所以 tf.matmul
L1 = tf.nn.tanh(Wx_plus_b_L1)

# 定义输出层
Weights_L2 = tf.Variable(tf.random_normal([10, 1]))
biases_L2 = tf.Variable(tf.zeros([1,1]))
Wx_plus_b_L2 = tf.matmul(L1, Weights_L2) + biases_L2  
prediction = tf.nn.tanh(Wx_plus_b_L2)

# 二次代价函数
loss = tf.reduce_mean(tf.square(y-prediction))
# 使用梯度下降法训练
train_step = tf.train.GradientDescentOptimizer(0.1).minimize(loss)

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for _ in range(2000):
#         print(sess.run(Weights_L1, feed_dict={x:x_data}))
#         print(sess.run(Wx_plus_b_L1, feed_dict={x:x_data}))
#         print(sess.run(Weights_L2, feed_dict={x:x_data}))
#         print(sess.run(Wx_plus_b_L2, feed_dict={x:x_data}))
        sess.run(train_step, feed_dict={x:x_data, y:y_data})
        
    # 获得预测值
    prediction_value = sess.run(prediction, feed_dict={x:x_data})
    
    # 画图
    plt.figure()
    plt.scatter(x_data, y_data)
    plt.plot(x_data, prediction_value, 'r', lw=5)
```
